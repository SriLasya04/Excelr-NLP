1. Create an IAM User

-Sign in to AWS Console

-Go to AWS IAM Console.

-Navigate to Users

-In the left navigation pane, click "Users" → "Add users".

-Set User Details

-Enter a username.

Select Access Type:

-AWS Management Console access (for logging into AWS Console).

-Programmatic access (for CLI, SDK, or API access).

Set Permissions

-Attach policies directly or add the user to an IAM Group.

-Set Tags (Optional)

-Add metadata (e.g., department, role).

-Review & Create User

-Review settings and click "Create user".

2. Create an IAM User Group
Navigate to User Groups

-In the IAM console, click "User groups" → "Create group".

Set Group Name

-Enter a name (e.g., Developers, Admins).

Attach Policies

-Select AWS Managed Policies (like AdministratorAccess, ReadOnlyAccess)

-Or create a Custom Policy.

-Add Users to the Group

-Choose users to add to this group.

Create the Group

-Click "Create group".

3. Best Practices
Use Groups for Permissions: Instead of assigning policies to users directly, assign them to groups.

-Follow the Principle of Least Privilege: Grant only necessary permissions.

-Enable MFA (Multi-Factor Authentication) for extra security.

-Rotate Access Keys regularly if using programmatic access.

Enable MFA for an IAM User (AWS Console)
Sign in to AWS Management Console

Go to AWS IAM Console.

Select the User

Click "Users" in the left menu.

Choose the user for whom you want to enable MFA.

Navigate to Security Credentials

Scroll down to the "Multi-Factor Authentication (MFA)" section.

Click "Manage MFA".

Choose MFA Type

Select "Virtual MFA device" (for apps like Google Authenticator, Authy, or AWS MFA).

Or "Hardware MFA device" (if using a physical key like YubiKey).

Set Up Virtual MFA

Click "Continue" and scan the QR code using an MFA app.

Enter two consecutive OTP codes generated by the app.

Confirm and Activate

Click "Assign MFA" to enable it for the user.
